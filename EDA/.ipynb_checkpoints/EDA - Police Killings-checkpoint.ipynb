```python
killings_by_state = killings['State'].value_counts()
```

```python
census = pd.read_csv('../US Census Data/nst-est2019-alldata.csv')
```

```python
census['Average pop'] = census['POPESTIMATE2013'] + census['POPESTIMATE2014'] + census['POPESTIMATE2015'] + census['POPESTIMATE2016'] + census['POPESTIMATE2017'] + census['POPESTIMATE2018'] + census['POPESTIMATE2019']
census['Average pop'] = census['Average pop'] / 7
```

```python
census.drop(census[census['STATE']==0].index, inplace=True)
```

```python
abbrevs = pd.read_csv('../US Census Data/csvData.csv')
abbrevs.head()
```

```python
census_merge = census.merge(abbrevs, left_on='NAME', right_on='State')
census_merge.head()
```

```python
new = census_merge[['Code', 'State', 'Average pop']]
new_merge = new.merge(killings_by_state, left_on='Code', right_on=killings_by_state.index)
new_merge['Killings per 100,000'] = new_merge['State_y'] / new_merge['Average pop'] * 100_000
new_merge_sorted = new_merge.sort_values('Killings per 100,000', ascending=True)
```

```python
fig,ax = plt.subplots(1,1,figsize=(10,10))
new_merge_sorted.plot.barh(x='State_x', y='Killings per 100,000', ax = ax)
```

```python
fig,ax = plt.subplots(1,1,figsize=(10,10))
new_merge_sorted[-10:].plot.barh(x='State_x', y='Killings per 100,000', ax = ax)
```

```python
states = states.to_crs("EPSG:3395")
```

```python
new_merge.head()
```

```python
states_merge = pd.merge(states, new_merge, left_on='STUSPS', right_on='Code')
states_merge.head()
```

```python
states_merge.shape
```

```python
#fig = plt.figure(figsize=(10,5))
states_merge = states_merge.to_crs("EPSG:3395")
fig, ax = plt.subplots(1,1, figsize=(10,5))
states_merge.boundary.plot(color='lightgrey', ax=ax)
states_merge.centroid.plot(cmap='spring',markersize=states_merge['Killings per 100,000']*50, ax = ax);
```

```python
us_boundary_map = states_merge.boundary.plot(figsize=(15,10), color='black')
states_merge.sort_values('Killings per 100,000', ascending=False)[0:10].plot(ax=us_boundary_map, color='red', hatch="///");
```

```python
fig, ax = plt.subplots(1,1, figsize=(15,10))
states_merge.boundary.plot(figsize=(15,10), color='lightgrey', ax=ax)
states_merge.plot(column='Killings per 100,000',ax=ax,legend=True,cmap='Reds',
                 legend_kwds={'label': 'Number of Police killings per 100,000',
                             'orientation':'horizontal'});
```
